## Hypothesis Testing for Linear Regression

Null hypthesis : cofficients of cement	slag,flyash,water,superplasticizer,coarseaggregate,fineaggregate and age = 0

# Splitting the data into Train-Validation-Test

```python
from sklearn.model_selection import train_test_split
#Train-test split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
#Train-validation split
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25)
```

## Standardizing the data for distance based algorithms like KNN and optimization based algorithm

```python
from sklearn.preprocessing import StandardScaler       #Scaling the data for  linear regression and KNN
scaler = StandardScaler()

X_train1 = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
X_valid1 = pd.DataFrame(scaler.transform(X_valid), columns = X_valid.columns)
X_test1= pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
```

# Model Building

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train1, y_train)
```

```python
y_pred = reg.predict(X_valid1)
```

```python
#MAE - Mean Absoulte Error
np.mean(np.abs(y_valid - y_pred))
```

```python
#MAPE
np.mean(np.abs(y_valid - y_pred)/y_valid)*100
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

reg_tree = DecisionTreeRegressor(max_depth=7)   #greedy algorithm
reg_tree.fit(X_train,y_train)
```

```python
y_pred = reg_tree.predict(X_valid)
```

```python
#MAE
np.mean(np.abs(y_valid - y_pred))
```

```python
#MAPE
np.mean(np.abs(y_valid - y_pred)/y_valid)*100
```

### Hyperparameter tuning - chosing best depth

```python
depth = np.arange(1,20)
mape=[]

for k in depth:
    reg_tree = DecisionTreeRegressor(max_depth=k)
    reg_tree.fit(X_train,y_train)
    y_pred = reg_tree.predict(X_valid)
    mape.append(np.mean(np.abs(np.mean(np.abs(y_valid - y_pred)/y_valid)*100)))
```

```python
plt.plot(depth, np.array(mape))
plt.show()
```

# KNN 

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=7)
```

```python
knn.fit(X_train1,y_train)
```

```python
y_pred = knn.predict(X_valid1)
```

```python
#MAE
np.mean(np.abs(y_valid - y_pred))
```

```python
#MAPE
np.mean(np.abs(y_valid - y_pred)/y_valid)*100
```

### Hyperparameter tuning - chosing best value of k

```python
k = np.arange(1,20)
mape=[]

for i in k:
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train1,y_train)
    y_pred = knn.predict(X_valid1)
    mape.append(np.mean(np.abs(np.mean(np.abs(y_valid - y_pred)/y_valid)*100)))
```

```python
plt.plot(k, np.array(mape))
plt.show()
```
